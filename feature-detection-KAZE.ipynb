```python
# Imports
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
```

```python
# Open and convert a input
# image from BGR to GRAYSCALE
image = cv.imread(filename = 'Figures/input.png',
                   flags = cv.IMREAD_GRAYSCALE)
```

```python
# KAZE is a feature detector and descriptor

# Initiate KAZE detector
KAZE = cv.KAZE_create()
```

```python
 # Find the keypoints with KAZE
keypoints = KAZE.detect(image, None)

# Print number of keypoints detected
print("Number of keypoints Detected:", len(keypoints), "\n")

# Compute the descriptors with KAZE
keypoints, descriptors = KAZE.compute(image, keypoints)

# Print the descriptor size in bytes
print("Size of Descriptor:", KAZE.descriptorSize(), "\n")

# Print the descriptor type
print("Type of Descriptor:", KAZE.descriptorType(), "\n")

# Print the default norm type
print("Default Norm Type:", KAZE.defaultNorm(), "\n")

# Print shape of descriptor
print("Shape of Descriptor:", descriptors.shape, "\n")
```

```python
# Draw only 50 keypoints on input image
image = cv.drawKeypoints(image = image,
                         keypoints = keypoints[:50],
                         outImage = None,
                         flags = cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
# Plot input image

# Turn interactive plotting off
plt.ioff()

# Create a new figure
plt.figure()
plt.axis('off')
plt.imshow(image)
plt.show()

plt.imsave(fname = 'Figures/feature-detection-KAZE.png',
           arr = image,
           dpi = 300)

# Close it
plt.close()
```
